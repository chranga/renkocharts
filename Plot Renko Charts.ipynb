# How to plot renko charts using Python
Reference: https://github.com/matplotlib/mplfinance/blob/master/examples/price-movement_plots.ipynb

Full guide: https://renkotraders.com/trading-matters/how-to-plot-renko-charts-in-python/

## Import Packages
Import pandas, pandas_datareader and mplfinance

```python
import pandas as pd
from pandas_datareader import data as pdr
import mplfinance as fplt
```

## Assign variables

```python
ticker = "MSFT"
sdate = "2020-01-01"
size = 2
df = pdr.get_data_yahoo(ticker, start = sdate)
```

## Output the dataframe
This gives you an idea of how the dataframe looks like

```python
df
```

## Plot renko chart using fixed brick size

```python
#filename = ticker.lower()+'.png'

fplt.plot(df,type='renko',renko_params=dict(brick_size=size),style='yahoo',figsize =(18,7), title = "RENKO CHART FOR {0}".format(ticker))
```

## Plot renko chart using an ATR value of 10
Change atr_length from 10 to any value of your choice

```python
fplt.plot(df,type='renko',renko_params=dict(brick_size='atr', atr_length=10), style='yahoo',figsize =(18,7), title = "RENKO CHART WITH ATR FOR {0}".format(ticker))
```

## Use all data and calculate ATR
Use atr_length = 'total'

```python
fplt.plot(df,type='renko',renko_params=dict(brick_size='atr', atr_length='total'),style='yahoo',figsize =(18,7), title = "RENKO CHART WITH ATR FOR {0}".format(ticker))
```
